```python
#sampling
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
xn =np.random.normal(45,10,100000) #45= mean #10 is the std deviation
plt.hist(xn,bins=100)
plt.show()
```

```python
#uniform distribution 
np.random.seed(10)
xu = np.random.uniform(10,100,100000)  #10 to 100 ia the range
plt.hist(xu,bins=100)
plt.show
```

```python
#we want a circle of x*x +y*y =100
x = np.random.uniform(-10,10,100)
yplus = np.sqrt(100-x*x)
yminus = - 1*(np.sqrt(100-x*x))
X = list(x)+ list(x)
Y =list(yplus)+list(yminus)
plt.figure(figsize=(6,6))

plt.scatter(X,Y,label="x^2+y^2=100")
plt.legend(loc ="center")
plt.show()
```
